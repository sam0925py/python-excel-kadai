```python


sheet = workbook.create_sheet('Sheet1')
sheet = workbook['Sheet1']


# Invoice title
sheet['B2'].value = '請 求 書'


# Company Information
sheet['B4'].value = "株式会社ABC"
sheet['B5'].value = "〒101-0022 東京都千代田区神田練塀町300"
sheet['B6'].value = "TEL:03-1234-5678 FAX:03-1234-5678"
sheet['B7'].value = "担当者名:鈴木一郎 様"

# Invoice Details
sheet['F4'].value = "No."
sheet['G4'].value = "0001"
sheet['F5'].value = "日付"
sheet['G5'].value = "2025/09/15"

# Item Table Header
#header = ['商品名',"数量", '単価', '金額']
#sheet.append(header) # This will be added starting from row 8 or 9 depending on previous content. Let's explicitly set the row.
sheet['B10'].value = '商品名'
sheet['C10'].value = '数量'
sheet['D10'].value = '単価'
sheet['E10'].value = '金額'

# Item Data
data = [['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000]]

# Append data starting from row 11
for row_index, row_data in enumerate(data):
    for col_index, value in enumerate(row_data):
        sheet.cell(row=11 + row_index, column=2 + col_index, value=value)

# Totals
total = sum(item[3] for item in data)
tax_rate = 0.10  # 10% tax rate
tax = total * tax_rate
total_with_tax = total + tax

sheet["B15"].value = "小計"
sheet["E15"].value = total
sheet["B16"].value = "消費税"
sheet["E16"].value = tax
sheet["B17"].value = "合計"
sheet["E17"].value = total_with_tax



```
