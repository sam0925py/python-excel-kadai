```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Step1: 2つのExcelファイルを読み込む
df_2022 = pd.read_excel('2022_年間売上表.xlsx')
df_2023 = pd.read_excel('2023_年間売上表.xlsx')

# Step2: データを連結
df_all = pd.concat([df_2022, df_2023], ignore_index=True)

# Step3: 「商品」「売上年」ごとに合計金額を集計
df_summary = df_all.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()

# Step4: 新しいExcelファイルに書き出し
output_file = '売上集計表.xlsx'
df_summary.to_excel(output_file, index=False)

# Step5: ヘッダー行に色を付ける（openpyxlで編集）
workbook = load_workbook(output_file)
ws = workbook.active

# ヘッダー行（1行目）を薄いグレー（#F2F2F2）に
header_fill = PatternFill(patternType='solid', fgColor='F2F2F2')
for row in ws.iter_rows(min_row=1, max_row=1, min_col=1, max_col=3):
    for cell in row:
        cell.fill = header_fill


workbook.save(output_file)
print("✅ 売上集計表.xlsx を作成しました")
```
